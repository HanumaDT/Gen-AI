```python
import os
from pptx import Presentation
import json
```

```python
ppt_dir = "Downloads/PPTs"
out_dir = "Downloads/Annotated_data"
os.makedirs(out_dir, exist_ok=True)
all_annotations = []
count = 1

# def get_rgb_colour(fill):
#     try:
#         if fill and fill.fore_color and fill.fore_color.rgb:
#             return tuple(fill.fore_color.rgb)  # Returns (R, G, B)
#         return None
#     except AttributeError:
#         return None

for filename in os.listdir(ppt_dir):
    count = count+1
    print(count)
    if filename.endswith(".pptx"):
        ppt_path = os.path.join(ppt_dir, filename)
        # print(ppt_path)
        prs = Presentation(ppt_path)
        annotations = []
        
        for slide_index, slide in enumerate(prs.slides):
            slide_data = {"slide_number": slide_index+1, "elements":[]}
            
            for shape in slide.shapes:
                shape_data = {
                    "shape_type": shape.shape_type,
                    "position": [shape.left, shape.top],
                    "size": [shape.width, shape.height]}

                # if shape_data['shape_type'] == 6 or 13 or 17:
                #     shape_data['fill_color'] = get_rgb_colour(shape.fill)
                
                if shape.has_text_frame:
                    shape_data['type'] = 'textbox'
                    shape_data['text'] = shape.text

                    # if shape.text_frame and shape.text_frame.text and shape.text_frame.paragraphs:
                    #     run = shape.text_frame.paragraphs[0].runs[0] if shape.text_frame.paragraphs[0].runs else None
                    #     if run and run.font and run.font.color and run.font.color.rgb:
                    #         shape_data['text_color'] = tuple(run.font.color.rgb)
                    #     else:
                    #         shape_data['text_color'] = None
                    
                elif shape.shape_type == 13:
                    image = shape.image
                    image_name = f"{filename.replace('.pptx', '')}_slide_{slide_index + 1}_image.jpg"
                    image_path = os.path.join(out_dir, image_name)
                    with open(image_path, "wb") as f:
                        f.write(image.blob)
                    shape_data['type'] = 'image'
                    shape_data['path'] = image_path
                    
                else:
                    shape_data['type'] = 'other'
                    
                slide_data["elements"].append(shape_data)

            annotations.append(slide_data)
            
        all_annotations.append({'filename': filename, 'annotations': annotations})

output_json_path = os.path.join(out_dir, "all_annotations.json")
with open(output_json_path, "w") as f:
    json.dump(all_annotations, f, indent = 4)

print("All the annotations are saved in all_annotations")
```
