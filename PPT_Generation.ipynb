```python
from google.colab import drive
drive.mount("/content/drive")
from google.colab import drive
from pptx import Presentation
from pptx.util import Inches
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN

# Function to convert hex to RGB
def hex_to_rgb(hex_color):
    hex_color = hex_color.lstrip('#')
    return RGBColor(int(hex_color[0:2], 16), int(hex_color[2:4], 16), int(hex_color[4:6], 16))

# Function to add slides
def add_slide(presentation, layout_type, title, content, background_color, text_color):
    slide = presentation.slides.add_slide(layout_type)

    # Set background color
    background = slide.background
    fill = background.fill
    fill.solid()
    fill.fore_color.rgb = hex_to_rgb(background_color)

    # Adjust title placeholder
    title_shape = slide.shapes.title
    if title_shape:
        title_shape.text = title
        title_frame = title_shape.text_frame
        title_frame.paragraphs[0].font.color.rgb = hex_to_rgb(text_color)
        title_frame.paragraphs[0].alignment = PP_ALIGN.CENTER
        title_shape.width = int(presentation.slide_width * 0.8)
        title_shape.height = Inches(1.5)
        title_shape.left = (presentation.slide_width - title_shape.width) // 2
        title_shape.top = Inches(0.5)

    # Adjust content placeholder
    if content:
        content_placeholder = slide.placeholders[1]  # Assuming the second placeholder is for content
        content_placeholder.text = content
        content_frame = content_placeholder.text_frame
        for paragraph in content_frame.paragraphs:
            paragraph.font.color.rgb = hex_to_rgb(text_color)
            paragraph.alignment = PP_ALIGN.LEFT

        # Dynamically set the size and position
        content_placeholder.width = int(presentation.slide_width * 0.9)
        content_placeholder.height = int(presentation.slide_height * 0.5)
        content_placeholder.left = (presentation.slide_width - content_placeholder.width) // 2
        content_placeholder.top = Inches(2)

# Create Presentation
presentation = Presentation()
presentation.slide_width = Inches(13.33)
presentation.slide_height = Inches(7.5)

# Define colors
colors = {
    "red": "#FF0000",
    "white": "#FFFFFF",
    "dark_gray": "#A9A9A9"
}

# Add Title Slide
add_slide(
    presentation,
    layout_type=presentation.slide_layouts[0],  # Title Slide Layout
    title="Presentation Title",
    content="Subtitle or Tagline",
    background_color=colors.get("red"),
    text_color=colors.get("dark_gray")
)

# Add Content Slide
add_slide(
    presentation,
    layout_type=presentation.slide_layouts[1],  # Content Slide Layout
    title="Section Title",
    content="- Main Point 1\n- Main Point 2\n- Main Point 3",
    background_color=colors.get("white"),
    text_color=colors.get("dark_gray")
)

# Save the Presentation
save_path = "/content/drive/MyDrive/Dynamic_Template_2.pptx"
presentation.save(save_path)
print(f"PowerPoint template created and saved as '{save_path}'")
```
